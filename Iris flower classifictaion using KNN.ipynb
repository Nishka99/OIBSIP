```python
# import dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Solving iris classification problem using KNN algorithm

```python
data=sns.load_dataset('iris')
```

```python
data
```

```python
data['species'].unique()
```

# independent variable

```python
data.columns
```

```python
X=data[['sepal_length','sepal_width','petal_length','petal_width']]
X
```

```python
X.drop([145],axis=0,inplace=True)
```

```python
X
```

# Dependent variable

```python
Y=data['species']
Y
```

```python
Y=Y.drop([145])
Y
```

```python
test=data.iloc[145]
```

```python
test
```

```python
X_test=test.iloc[:-1]
X_test
```

```python
Y_test=test.iloc[-1]
Y_test
```

# Assume K value

```python
K=4
```

```python
distance=[]
for X_train in X.itertuples():
    s=0
    for i,j in zip(list(X_train)[1:],X_test):
        s=s+(i-j)**2
    d=np.sqrt(s)
    distance.append(d)
distance
```

```python
X['species']=Y
X['Distance']=distance
```

```python
X
```

```python
X_sort=X.sort_values(by='Distance')
X_sort
```

```python
(X_sort['species'].iloc[:K]).mode()
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data.columns
```

```python
X=data[['sepal_length','sepal_width','petal_length','petal_width']]
Y=data['species']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=10)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn
```

```python
knn.fit(X_train,Y_train)
```

# Prediction for training dataset

```python
train_pred=knn.predict(X_train)
```

```python
pd.DataFrame({'actual_species':Y_train,'predicted_species':train_pred})
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_train,train_pred)
```

```python
knn.score(X_train,Y_train)
```

# Prediction for testing dataset

```python
test_pred=knn.predict(X_test)
```

```python
pd.DataFrame({'actual_species':Y_test,'predicted_species':test_pred})
```

```python
confusion_matrix(Y_test,test_pred)
```

```python
knn.score(X_test,Y_test)
```
